```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.random.rand(30)
data
```

```python
plt.plot(data)
```

```python
plt.scatter(data,data)
plt.scatter(data[:15],data[15:])
plt.show()
plt.scatter(data[:15],data[15:],color='orange')
plt.show()
```

```python
plt.hist(data)
plt.show()
plt.hist(data,bins=30)
plt.text(0.5, 0.0, 'np.random.rand(5000)-10.0')
plt.show()
```

```python
plt.bar(data,data)
```
